```python
seconds_per_hour * 24
```

There are 86,400 seconds in a day.

```python
seconds_per_day = seconds_per_hour * 24
seconds_per_day
```

The variable seconds_per_day stores the value 86400.


```python
seconds_per_day / seconds_per_hour
```

The answer is 24.0. This means there are 24 hours in a day. The decimal appears because floating-point division (/) was used.

```python
seconds_per_day // seconds_per_hour
```

The answer is 24. Yes, it agrees with the floating-point value from question 3.5 except that integer division (//) does not include the decimal .0.
